### Exploratory data analysis (EDA):
- Exploratory data analysis (EDA) is used by data scientists to analyze and investigate data sets and summarize their main characteristics, often employing data visualization methods. ... It can also help determine if the statistical techniques you are considering for data analysis are appropriate. (source google 02-02-2021)

### Univariate Analysis:
- Univariate analysis is the simplest form of data analysis where the data being analyzed contains only one variable. (source google 02-02-2021)

### Bivariate Analysis:
- Bivariate analysis is one of the simplest forms of quantitative (statistical) analysis. It involves the analysis of two variables (often denoted as X, Y), for the purpose of determining the empirical relationship between them. (source google 02-02-2021)

### MultiVariate Analysis:
- Multivariate analysis is a set of statistical techniques used for analysis of data that contain more than one variable. (source google 02-02-2021)

# **Univariate Analysis**

```
df_1 = df.loc[df['Outcome'] == 1]
df_1
```

```
df_0 = df.loc[df['Outcome'] == 0]
df_0
```

```
#Pregnancies
plt.plot(df_1['Pregnancies'], np.zeros_like(df_1['Pregnancies']), '*')
plt.plot(df_0['Pregnancies'], np.zeros_like(df_0['Pregnancies']), 'o')
plt.show()
```

```
#Glucose
plt.plot(df_1['Glucose'], np.zeros_like(df_1['Glucose']), '*')
plt.plot(df_0['Glucose'], np.zeros_like(df_0['Glucose']), 'o')
plt.show()
```

```
#BloodPressure
plt.plot(df_1['BloodPressure'], np.zeros_like(df_1['BloodPressure']), '*')
plt.plot(df_0['BloodPressure'], np.zeros_like(df_0['BloodPressure']), 'o')
plt.show()
```

```
#SkinThickness
plt.plot(df_1['SkinThickness'], np.zeros_like(df_1['SkinThickness']), '*')
plt.plot(df_0['SkinThickness'], np.zeros_like(df_0['SkinThickness']), 'o')
plt.show()
```

```
#Insulin
plt.plot(df_1['Insulin'], np.zeros_like(df_1['Insulin']), '*')
plt.plot(df_0['Insulin'], np.zeros_like(df_0['Insulin']), 'o')
plt.show()
```

```
#BMI
plt.plot(df_1['BMI'], np.zeros_like(df_1['BMI']), '*')
plt.plot(df_0['BMI'], np.zeros_like(df_0['BMI']), 'o')
plt.show()
```

```
#	DiabetesPedigreeFunction
plt.plot(df_1['DiabetesPedigreeFunction'], np.zeros_like(df_1['DiabetesPedigreeFunction']), '*')
plt.plot(df_0['DiabetesPedigreeFunction'], np.zeros_like(df_0['DiabetesPedigreeFunction']), 'o')
plt.show()
```

```
#	Age
plt.plot(df_1['Age'], np.zeros_like(df_1['Age']), '*')
plt.plot(df_0['Age'], np.zeros_like(df_0['Age']), 'o')
plt.xlabel('')
plt.show()
```

# **Bivariate Analysis**

- **FacetGrid** object takes a dataframe as input and the names of the variables that will form the row, column, or hue dimensions of the grid. The variables should be categorical and the data at each level of the variable will be used for a facet along that axis. (source google 02-02-2021)

```
sns.FacetGrid(df, hue = 'Outcome', size=5).map(plt.scatter,"Glucose","Pregnancies").add_legend();
plt.show()
```

```
sns.FacetGrid(df, hue = 'Outcome', size=5).map(plt.scatter,"BloodPressure","SkinThickness").add_legend();
plt.show()
```

```
sns.FacetGrid(df, hue = 'Outcome', size=5).map(plt.scatter,"Insulin","BMI").add_legend();
plt.show()
```

```
sns.FacetGrid(df, hue = 'Outcome', size=5).map(plt.scatter,"DiabetesPedigreeFunction","Age").add_legend();
plt.show()
```
